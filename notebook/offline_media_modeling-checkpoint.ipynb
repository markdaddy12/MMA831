```python
# === ALL-IN-ONE: Offline Media MMM with Adstock + Nonlinear Curves ===
# Fresh run: defines everything, fits models, saves summary, and plots one example.

import os
import pandas as pd
import numpy as np
import re
from pathlib import Path
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# ----------------------------
# 0) Load data
# ----------------------------
DATA_PATH = '../data/cleaned_globestay.parquet'
df = pd.read_parquet(DATA_PATH)
df.columns = (
    df.columns
      .str.strip()
      .str.replace(r"[^\w]+", "_", regex=True)
      .str.lower()
)

# ----------------------------
# 1) Country / Time config (you confirmed these)
# ----------------------------
COUNTRY = "country"
DATECOL = "week_start"
if DATECOL in df.columns and df[DATECOL].dtype == "object":
    df[DATECOL] = pd.to_datetime(df[DATECOL], errors="coerce")
if DATECOL in df.columns and (np.issubdtype(df[DATECOL].dtype, np.datetime64) or np.issubdtype(df[DATECOL].dtype, np.number)):
    df = df.sort_values([COUNTRY, DATECOL])

print("COUNTRIES:", df[COUNTRY].value_counts(dropna=False).to_dict())

# ----------------------------
# 2) Pick target and detect offline spend columns
# ----------------------------
sales_cols = [c for c in df.columns if c.startswith("sales_")]
if not sales_cols:
    raise KeyError("No 'sales_*' columns found. Inspect df.columns and set TARGET_COL manually.")
df["sales_total"] = df[sales_cols].sum(axis=1)
TARGET_COL = "sales_total"

channel_keywords = {
    "tv":    [r"(^|_)tv($|_)", r"brandtv", r"brand_tv", r"tv_brand", r"brand_tv_grps?", r"tv_.*cost", r"cost_.*tv", r"spend_.*tv", r"tv_.*spend"],
    "radio": [r"(^|_)radio($|_)", r"rad_?cost", r"cost_.*radio", r"spend_.*radio"],
    "print": [r"(^|_)print($|_)", r"newspaper", r"mag(azine)?", r"cost_.*print", r"spend_.*print"],
    "ooh":   [r"(^|_)ooh($|_)", r"out_?of_?home", r"outdoor", r"cost_.*ooh", r"spend_.*ooh"]
}

def find_first_matching_column(regex_list, columns):
    for rgx in regex_list:
        pat = re.compile(rgx)
        # prioritize spend/cost/net
        for col in columns:
            if pat.search(col) and any(k in col for k in ["cost","spend","net"]):
                return col
        for col in columns:
            if pat.search(col):
                return col
    return None

offline_cols = {ch: find_first_matching_column(pats, df.columns.tolist())
                for ch, pats in channel_keywords.items()}

print("\nDETECTED OFFLINE COLUMNS:")
for ch, col in offline_cols.items():
    print(f"{ch.upper():5s} -> {col}")

# If any missing, try to guess from column list to help you adjust
if any(v is None for v in offline_cols.values()):
    print("\nHINT: Candidate columns in your file (first 120):")
    print([c for c in list(df.columns)[:120] if any(k in c for k in ["tv","radio","print","ooh","cost","spend","net"])])

# -----------------------------------
# 3) Adstock transform
# -----------------------------------
def adstock_transform(x, alpha):
    x = np.nan_to_num(np.asarray(x, dtype=float))
    S = np.zeros_like(x)
    if len(S) == 0: 
        return S
    S[0] = x[0]
    for t in range(1, len(x)):
        S[t] = x[t] + alpha * S[t-1]
    return S

adstock_alpha = {"tv": 0.85, "radio": 0.6, "print": 0.5, "ooh": 0.7}

# Build adstock columns per available channel
for ch, spend_col in offline_cols.items():
    if spend_col is None:
        continue
    ad_col = f"{ch}_adstock"
    df[ad_col] = np.nan
    for g, dsub in df.groupby(COUNTRY):
        idx = dsub.index
        df.loc[idx, ad_col] = adstock_transform(dsub[spend_col].values, adstock_alpha[ch])

print("\nADSTOCKS:", [c for c in df.columns if c.endswith("_adstock")])

# -----------------------------------
# 4) Response functions and helpers
# -----------------------------------
def logistic(S, L, U, k, x0):
    return L + (U - L) / (1.0 + np.exp(-k * (S - x0)))

def gompertz(S, a, b, c, m):
    return a + (b - a) * np.exp(-c * np.exp(-(S - m)))

def adbudg(S, a, b, c, d):
    S = np.maximum(S, 0)
    return b + (a - b) * ((S**c) / (S**c + d))

def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - ss_res/ss_tot if ss_tot > 0 else np.nan

from scipy.optimize import curve_fit
def fit_and_summarize(x, y, model_name):
    if model_name == "logistic":
        L0, U0 = np.percentile(y, 5), np.percentile(y, 95)
        k0, x0_0 = 1e-4, np.median(x)
        p0 = [L0, U0, k0, x0_0]
        bounds = ([-np.inf, -np.inf, -np.inf, -np.inf], [np.inf, np.inf, np.inf, np.inf])
        f = logistic
    elif model_name == "gompertz":
        a0, b0 = np.percentile(y, 5), np.percentile(y, 95)
        c0, m0 = 0.01, np.median(x)
        p0 = [a0, b0, c0, m0]
        bounds = ([-np.inf, -np.inf, 1e-6, -np.inf], [np.inf, np.inf, np.inf, np.inf])
        f = gompertz
    elif model_name == "adbudg":
        a0, b0 = np.percentile(y, 95), np.percentile(y, 5)
        c0, d0 = 1.0, max(np.percentile(x, 75), 1e-6)
        p0 = [a0, b0, c0, d0]
        bounds = ([0, -np.inf, 1e-6, 1e-8], [np.inf, np.inf, 10, np.inf])
        f = adbudg
    else:
        raise ValueError(model_name)
    try:
        params, _ = curve_fit(f, x, y, p0=p0, bounds=bounds, maxfev=20000)
        yhat = f(x, *params)
        r2  = r2_score(y, yhat)
        return params, yhat, r2, f
    except Exception:
        return None, None, np.nan, None

# -----------------------------------
# 5) Fit per country × channel and summarize
# -----------------------------------
summary_rows = []
models = ["adbudg", "logistic", "gompertz"]

for ch in ["tv","radio","print","ooh"]:
    ad_col = f"{ch}_adstock"
    if ad_col not in df.columns:
        continue
    for g, dsub in df.dropna(subset=[ad_col, TARGET_COL]).groupby(COUNTRY):
        x = dsub[ad_col].values.astype(float)
        y = dsub[TARGET_COL].values.astype(float)
        if len(x) < 20 or np.nanstd(x) == 0:
            continue
        best = None
        for m in models:
            params, yhat, r2, f = fit_and_summarize(x, y, m)
            if params is None:
                continue
            if (best is None) or (r2 > best["r2"]):
                best = {"model": m, "params": params, "r2": r2, "f": f}
        if best is None:
            continue

        S_now = np.mean(x); Y_now = np.mean(y)
        f = best["f"]; params = best["params"]
        h = max(1e-6, S_now*1e-4)
        mr = (f(S_now + h, *params) - f(S_now, *params)) / h
        elasticity = mr * (S_now / Y_now) if Y_now > 0 else np.nan
        roi_avg = (f(S_now, *params) - np.percentile(y, 5)) / max(S_now, 1e-6)

        summary_rows.append({
            "country": g,
            "channel": ch,
            "best_model": best["model"],
            "r2": round(best["r2"], 4),
            "S_mean": S_now,
            "Y_mean": Y_now,
            "marginal_return": mr,
            "elasticity": elasticity,
            "avg_roi_proxy": roi_avg
        })

summary = pd.DataFrame(summary_rows).sort_values(["country","channel"])
print("\n=== MODEL SUMMARY (first rows) ===")
try:
    from IPython.display import display
    display(summary.head(12))
except Exception:
    print(summary.head(12))

OUTPUT_DIR = '../output/offline_channels/'
os.makedirs(OUTPUT_DIR, exist_ok=True)
out_path = Path(OUTPUT_DIR) / "offline_media_model_summary.csv"
summary.to_csv(out_path, index=False)
print(f"\nSaved summary to: {out_path.resolve()}")

# -----------------------------------
# 6) Response plot helper
# -----------------------------------
def plot_response_example(country, channel):
    ch = channel.lower()
    ad_col = f"{ch}_adstock"
    dsub = df[df[COUNTRY]==country].dropna(subset=[ad_col, TARGET_COL]).copy()
    if dsub.empty:
        print(f"No data for {country}-{channel}")
        return
    x = dsub[ad_col].values.astype(float)
    y = dsub[TARGET_COL].values.astype(float)
    best = None
    for m in models:
        params, yhat, r2, f = fit_and_summarize(x, y, m)
        if params is None:
            continue
        if (best is None) or (r2 > best["r2"]):
            best = {"model": m, "params": params, "r2": r2, "f": f}
    if best is None:
        print("Could not fit model.")
        return
    xx = np.linspace(x.min(), x.max(), 200)
    yy = best["f"](xx, *best["params"])
    plt.figure(figsize=(6,4))
    plt.scatter(x, y, s=10, alpha=0.4, label="weekly data")
    plt.plot(xx, yy, linewidth=2, label=f"{best['model']} fit (R²={best['r2']:.2f})")
    plt.title(f"{country} – {channel.upper()} response")
    plt.xlabel(f"{channel.upper()} adstock")
    plt.ylabel(TARGET_COL)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- Quick sanity plot (change labels if needed) ---
plot_response_example("us","tv")
```

```python
# ==== NUMERIC STABILITY PATCH: safe exp + per-fit scaling ====

import numpy as np
from scipy.optimize import curve_fit

# 1) Safe exp to avoid overflow
def _safe_exp(z):
    # Clip z to a reasonable window; 40 ~ exp(40) ≈ 2.35e17
    return np.exp(np.clip(z, -40, 40))

# 2) Response functions on a SCALED x (x_scaled = x / s_scale)
def logistic_scaled(x_scaled, L, U, k, x0_scaled):
    # y = L + (U-L) / (1 + exp(-k (x_scaled - x0_scaled)))
    return L + (U - L) / (1.0 + _safe_exp(-k * (x_scaled - x0_scaled)))

def gompertz_scaled(x_scaled, a, b, c, m_scaled):
    # y = a + (b-a) * exp( -c * exp( -(x_scaled - m_scaled) ) )
    return a + (b - a) * _safe_exp(-c * _safe_exp(-(x_scaled - m_scaled)))

def adbudg_scaled(x_scaled, a, b, c, d_scaled):
    # y = b + (a-b) * (x_scaled^c / (x_scaled^c + d_scaled))
    x_scaled = np.maximum(x_scaled, 0)
    num = np.power(x_scaled, c)
    den = num + d_scaled
    return b + (a - b) * (num / np.maximum(den, 1e-12))

def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - ss_res/ss_tot if ss_tot > 0 else np.nan

def _fit_one_model(x, y, model_name):
    """
    Fits model with per-fit scaling for x (adstock).
    Returns dict with model name, params, r2, scaler, f_predict(x) on ORIGINAL scale.
    """
    # scale x so its 95th percentile ~ 1 (or max if tiny)
    s_scale = np.percentile(x, 95)
    if not np.isfinite(s_scale) or s_scale <= 0:
        s_scale = np.max(x) if np.max(x) > 0 else 1.0
    x_scaled = x / s_scale

    if model_name == "logistic":
        L0, U0 = np.percentile(y, 5), np.percentile(y, 95)
        p0 = [L0, U0, 1.0, np.median(x_scaled)]  # k starts at 1.0 on scaled x
        bounds = ([-np.inf, -np.inf, 1e-6, -10], [np.inf, np.inf, 100.0, 10])
        f_scaled = logistic_scaled
    elif model_name == "gompertz":
        a0, b0 = np.percentile(y, 5), np.percentile(y, 95)
        p0 = [a0, b0, 1.0, np.median(x_scaled)]
        bounds = ([-np.inf, -np.inf, 1e-6, -10], [np.inf, np.inf, 100.0, 10])
        f_scaled = gompertz_scaled
    elif model_name == "adbudg":
        a0, b0 = np.percentile(y, 95), np.percentile(y, 5)
        p0 = [a0, b0, 1.0, 0.5]  # d_scaled around 0.5
        bounds = ([0, -np.inf, 1e-4, 1e-6], [np.inf, np.inf, 10.0, 100.0])
        f_scaled = adbudg_scaled
    else:
        raise ValueError(model_name)

    try:
        params, _ = curve_fit(f_scaled, x_scaled, y, p0=p0, bounds=bounds, maxfev=40000)
        yhat = f_scaled(x_scaled, *params)
        r2 = r2_score(y, yhat)
    except Exception:
        return None

    # prediction function on ORIGINAL x
    def f_orig(x_input):
        xs = (np.asarray(x_input, dtype=float) / s_scale)
        return f_scaled(xs, *params)

    # derivative on ORIGINAL x via small-step delta (stable)
    def fprime_orig(x0):
        h = max(1e-6, (x0 if x0 > 0 else 1.0) * 1e-4)
        return (f_orig(x0 + h) - f_orig(x0)) / h

    return {
        "model": model_name,
        "params": params,
        "r2": r2,
        "s_scale": s_scale,
        "f": f_orig,
        "fprime": fprime_orig
    }

def fit_with_stability(x, y, models=("adbudg","logistic","gompertz")):
    best = None
    for m in models:
        res = _fit_one_model(x, y, m)
        if res is None: 
            continue
        if (best is None) or (res["r2"] > best["r2"]):
            best = res
    return best

# === Refit summary with stable fitting ===
stable_rows = []
for ch in ["tv","radio","print","ooh"]:
    ad_col = f"{ch}_adstock"
    if ad_col not in df.columns:
        continue
    for g, dsub in df.dropna(subset=[ad_col, TARGET_COL]).groupby(COUNTRY):
        x = dsub[ad_col].to_numpy(dtype=float)
        y = dsub[TARGET_COL].to_numpy(dtype=float)
        if len(x) < 20 or np.nanstd(x) == 0:
            continue
        best = fit_with_stability(x, y)
        if best is None:
            continue
        S_now = float(np.mean(x))
        Y_now = float(np.mean(y))
        mr = float(best["fprime"](S_now))                 # marginal return on ORIGINAL x scale
        elasticity = mr * (S_now / Y_now) if Y_now > 0 else np.nan
        # avg ROI proxy (relative to low sales baseline)
        roi_avg = (float(best["f"](S_now)) - np.percentile(y, 5)) / max(S_now, 1e-6)

        stable_rows.append({
            "country": g,
            "channel": ch,
            "best_model": best["model"],
            "r2": round(best["r2"], 4),
            "S_mean": S_now,
            "Y_mean": Y_now,
            "marginal_return": mr,
            "elasticity": elasticity,
            "avg_roi_proxy": roi_avg
        })

summary_stable = pd.DataFrame(stable_rows).sort_values(["country","channel"])
print("\n=== STABLE MODEL SUMMARY (first rows) ===")
try:
    from IPython.display import display
    display(summary_stable.head(12))
except:
    print(summary_stable.head(12))

summary_stable.to_csv(Path(OUTPUT_DIR) / "offline_media_model_summary_stable.csv", index=False)
print("Saved: offline_media_model_summary_stable.csv")

# Override the plotter to use stable fit
def plot_response_example(country, channel):
    ch = channel.lower()
    ad_col = f"{ch}_adstock"
    dsub = df[df[COUNTRY]==country].dropna(subset=[ad_col, TARGET_COL]).copy()
    if dsub.empty:
        print(f"No data for {country}-{channel}")
        return
    x = dsub[ad_col].to_numpy(dtype=float)
    y = dsub[TARGET_COL].to_numpy(dtype=float)
    best = fit_with_stability(x, y)
    if best is None:
        print("Could not fit model.")
        return
    xx = np.linspace(x.min(), x.max(), 200)
    yy = best["f"](xx)
    plt.figure(figsize=(6,4))
    plt.scatter(x, y, s=10, alpha=0.35, label="weekly data")
    plt.plot(xx, yy, linewidth=2, label=f"{best['model']} fit (R²={best['r2']:.2f})")
    plt.title(f"{country} – {channel.upper()} response")
    plt.xlabel(f"{channel.upper()} adstock")
    plt.ylabel(TARGET_COL)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
